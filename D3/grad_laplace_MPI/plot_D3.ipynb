```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure()

data1 = np.loadtxt("data/weak_scaling1.dat")
data2 = np.loadtxt("data/weak_scaling2.dat")

size1 = data1[:,0]
msize1 = data1[:,2]
time1 = data1[:,3] 

size2 = data2[:,0]
msize2 = data2[:,2]
time2 = data2[:,3] 

plt.plot(size1, time1, label = "Msize = 1280000")
plt.plot(size2, time2, label = "Msize = 2560000")

plt.title("Weak Scaling: CondNumb = $10^5$ and $\hat{r} = 10^{-16}$")
plt.xlabel("NPE")
plt.ylabel("$time$")
plt.legend()
plt.savefig("data/weak_scaling.png")

plt.show()
```

```python
plt.figure()

data1 = np.loadtxt("data/strong_scaling1.dat")

serial = data1[0,1]
size = data1[:,0]
time = data1[:,1]



plt.plot(size, serial/time, label = "Msize = 500000")

plt.title("Strong Scaling: CondNumb = $10^5$ and $\hat{r} = 10^{-20}$")
plt.xlabel("NPE")
plt.ylabel("$time$")
plt.legend()
plt.savefig("data/strong_scaling.png")

plt.show()
```
